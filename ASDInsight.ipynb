```python
df.drop('Ethnicity', axis=1, inplace=True)
```

```python
df.drop('Who_completed_the_test', axis=1, inplace=True)
```

```python
cat=list(df.select_dtypes(include=['object']))
df[cat]
```

```python
for i in cat:
  print(df[i].value_counts(),'\n')
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in cat:
  df[i]=l.fit_transform(df[i])
```

```python
df[cat]
```

```python
X=df.drop('ASD_traits', axis=1)
y=df['ASD_traits']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
y_train
```

```python
#svm
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
```

```python
model
```

```python
import numpy as np
```

```python
y_preds=model.predict(X_test)
y_preds
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score1=accuracy_score(y_test, y_preds)
score1
```

```python
print(classification_report(y_test,y_preds))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
```

```python
y_preds2=model2.predict(X_test)
y_preds2
```

```python
score2=accuracy_score(y_test,y_preds2)
score2
```

```python
print(classification_report(y_test,y_preds2))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()
model3.fit(X_train,y_train)
```

```python
y_preds3=model3.predict(X_test)
y_preds3
```

```python
score3=accuracy_score(y_test, y_preds3)
score3
```

```python
#classification report
print(classification_report(y_test,y_preds3))
```

```python
from sklearn.linear_model import LogisticRegression
model4=LogisticRegression()
model4.fit(X_train,y_train)
```

```python
y_preds4=model4.predict(X_test)
y_preds4
```

```python
score4=accuracy_score(y_test, y_preds4)
score4
```

```python
print(classification_report(y_test,y_preds4))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model5=KNeighborsClassifier()
model5.fit(X_train,y_train)
```

```python
y_preds5=model5.predict(X_test)
y_preds5
```

```python
score5=accuracy_score(y_test, y_preds5)
score5
```

```python
from sklearn.ensemble import VotingClassifier
model6=VotingClassifier(estimators=[('svm',model),('dt',model2),('rf',model3),('lr',model4),('knn',model5)],voting='hard')
```

```python
model6.fit(X_train,y_train)
```

```python
y_preds6=model6.predict(X_test)
y_preds6
```

```python
score6=accuracy_score(y_test,y_preds6)
score6
```

```python
print(classification_report(y_test,y_preds6))
```
